```python
def is_item_set_frequent(itemset, support_percent):
    data = pd.read_csv('associationruletestdata.txt', delimiter="\t", header=None)

    i = 1
    n_columns = len(data.columns)

    for column in data:
        s = 'G' + str(i) + '_'
        if i < n_columns:
            data[column] = s + data[column].astype(str)
        i = i + 1

    items = itemset.split(",")
    cnt = 0

    total_transactions = data.shape[0]
    support_threshold = (support_percent/100) * total_transactions

    # Iterate over each row
    for i in range((data.shape[0])):
        row_list = list(data.iloc[i, :])
        if set(items).issubset(set(row_list)):
            cnt = cnt + 1

    print(cnt)
    if cnt >= support_threshold:
        return True
    else:
        return False


count = is_item_set_frequent("G95_Down", 30)
print(count)
```
